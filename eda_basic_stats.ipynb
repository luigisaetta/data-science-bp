### How to get the basic statistics for a dataset

In this example I'm going to show how, with few lines of code, you can get and save the basic stats for an entire dataset.

For this example we will be using:

https://objectstorage.us-ashburn-1.oraclecloud.com/n/bigdatadatasciencelarge/b/hosted-ds-datasets/o/synthetic%2Forcl_attrition.csv

```python
import pandas as pd
```

```python
URL = "https://objectstorage.us-ashburn-1.oraclecloud.com/n/bigdatadatasciencelarge/b/hosted-ds-datasets/o/synthetic%2Forcl_attrition.csv"

data_orig = pd.read_csv(URL)

data_orig.head()
```

```python
stats = data_orig.describe().T

# stats is a DataFrame, so if you want to process to improve visualization, it is easy
stats
```

```python
# for example, you maybe want to save the stats. This way if you get a new version, with more data , of the dataset you can check if stats have changed significantly
```

```python
# save for later usage
stats.to_csv("stats1.csv")
```
